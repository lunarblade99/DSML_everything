```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)
# a*b won't work because it perform ele wise ops

print(a,'\n\n',b)
```

```python
a @ b #same as matmul operation
```

```python
np.dot(a,b) #same as matmul operation
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.dot(a,b) #dot product -> 1*4 + 2*5 + 3*6 = 32
```
