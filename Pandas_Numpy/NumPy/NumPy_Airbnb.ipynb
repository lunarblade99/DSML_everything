```python
#promoters(9-10) & detractors(0-6) [neutrals(7-8)]
```

```python
detractors = len(score[score<=6])
detractors
```

```python
total = len(score)
```

```python
detractors_percent = (detractors/total)*100
detractors_percent
```

```python
promoters = len(score[score>=9])
promoters
```

```python
promoters_percent = (promoters/total)*100
promoters_percent
```

```python
NPS = promoters_percent - detractors_percent
NPS
```

```python
#let's make this arr of int as categorical arr
cat_arr = np.empty(dtype='str', shape=score.shape)
cat_arr
```

```python
cat_arr[score>=9] = 'P'
cat_arr[(score<9)&(score>=7)] = 'N'
cat_arr[score<=6] = 'D'
cat_arr
```
