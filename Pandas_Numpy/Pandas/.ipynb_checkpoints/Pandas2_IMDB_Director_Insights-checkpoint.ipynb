```python
movies = pd.read_csv("movies.csv", index_col=0)
movies.reset_index(drop=True, inplace=True)
movies.head()
```

```python
data = movies.merge(directors, how='left', left_on='director_id',right_on='id')
```

```python
data.drop(['director_id','id_y'],axis=1,inplace=True)
data.head(3)
```
